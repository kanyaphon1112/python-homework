````python
student_info = "‡∏£‡∏´‡∏±‡∏™:6601234|‡∏ä‡∏∑‡πà‡∏≠:‡∏™‡∏°‡∏ä‡∏≤‡∏¢ ‡πÉ‡∏à‡∏î‡∏µ|‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô:85"

parts = student_info.split('|')


student_id = parts[0].split(':')[1]
name = parts[1].split(':')[1]
score = int(parts[2].split(':')[1])

print(f"‡∏£‡∏´‡∏±‡∏™‡∏ô‡∏±‡∏Å‡∏®‡∏∂‡∏Å‡∏©‡∏≤: {student_id}")
print(f"‡∏ä‡∏∑‡πà‡∏≠: {name}")
print(f"‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô: {score} (‡πÄ‡∏õ‡πá‡∏ô {type(score).__name__})")

# ‡∏û‡∏¥‡∏°‡∏û‡πå‡πÇ‡∏Ñ‡πâ‡∏î‡∏Ñ‡∏≥‡∏ï‡∏≠‡∏ö‡∏ó‡∏µ‡πà‡∏ô‡∏µ‡πà
````

```
products = ["‡∏ô‡∏°", "‡∏Ç‡∏ô‡∏°‡∏õ‡∏±‡∏á", "‡πÑ‡∏Ç‡πà", "‡∏ô‡πâ‡∏≥‡∏ï‡∏≤‡∏•"]
```

‡πÉ‡∏´‡πâ‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô‡πÇ‡∏õ‡∏£‡πÅ‡∏Å‡∏£‡∏°‡πÄ‡∏û‡∏∑‡πà‡∏≠:

2. ‡πÅ‡∏ó‡∏£‡∏Å "‡∏ä‡∏≤" ‡πÉ‡∏ô‡∏ï‡∏≥‡πÅ‡∏´‡∏ô‡πà‡∏á‡∏ó‡∏µ‡πà 2 (‡πÉ‡∏ä‡πâ insert)

4. ‡πÅ‡∏™‡∏î‡∏á‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î
5. ‡πÅ‡∏™‡∏î‡∏á‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î
6. ‡πÅ‡∏™‡∏î‡∏á‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤ 3 ‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£‡πÅ‡∏£‡∏Å

**Expected Output:**
```
‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤: ['‡∏ô‡∏°', '‡∏Ç‡∏ô‡∏°‡∏õ‡∏±‡∏á', '‡∏ä‡∏≤', '‡∏ô‡πâ‡∏≥‡∏ï‡∏≤‡∏•', '‡πÄ‡∏ô‡∏¢', '‡πÅ‡∏¢‡∏°']
‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î: 6 ‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£
‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤ 3 ‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£‡πÅ‡∏£‡∏Å: ['‡∏ô‡∏°', '‡∏Ç‡∏ô‡∏°‡∏õ‡∏±‡∏á', '‡∏ä‡∏≤']
```

### ‡∏Ç‡πâ‡∏≠ 3: ‡∏Å‡∏≤‡∏£‡πÉ‡∏ä‡πâ‡∏á‡∏≤‡∏ô Dictionary (Student Records)

‡πÉ‡∏´‡πâ‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô‡πÇ‡∏õ‡∏£‡πÅ‡∏Å‡∏£‡∏°‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏à‡∏±‡∏î‡∏Å‡∏≤‡∏£‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ô‡∏±‡∏Å‡∏®‡∏∂‡∏Å‡∏©‡∏≤:
1. ‡∏™‡∏£‡πâ‡∏≤‡∏á dictionary ‡∏Ç‡∏≠‡∏á‡∏ô‡∏±‡∏Å‡∏®‡∏∂‡∏Å‡∏©‡∏≤ 3 ‡∏Ñ‡∏ô ‡πÇ‡∏î‡∏¢‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏Ñ‡∏ô‡∏°‡∏µ ‡∏£‡∏´‡∏±‡∏™, ‡∏ä‡∏∑‡πà‡∏≠, ‡πÅ‡∏•‡∏∞ ‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô
2. ‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏ô‡∏±‡∏Å‡∏®‡∏∂‡∏Å‡∏©‡∏≤‡∏Ñ‡∏ô‡∏ó‡∏µ‡πà 4 ‡∏•‡∏á‡πÉ‡∏ô dictionary
3. ‡∏≠‡∏±‡∏õ‡πÄ‡∏î‡∏ï‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô‡∏Ç‡∏≠‡∏á‡∏ô‡∏±‡∏Å‡∏®‡∏∂‡∏Å‡∏©‡∏≤‡∏Ñ‡∏ô‡πÉ‡∏î‡∏Ñ‡∏ô‡∏´‡∏ô‡∏∂‡πà‡∏á
4. ‡πÅ‡∏™‡∏î‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ô‡∏±‡∏Å‡∏®‡∏∂‡∏Å‡∏©‡∏≤‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î (‡πÉ‡∏´‡πâ‡πÉ‡∏ä‡πâ loop)
5. ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢‡∏Ç‡∏≠‡∏á‡∏ô‡∏±‡∏Å‡∏®‡∏∂‡∏Å‡∏©‡∏≤‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î

**Expected Output:**
```
‡∏ô‡∏±‡∏Å‡∏®‡∏∂‡∏Å‡∏©‡∏≤‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î: 4 ‡∏Ñ‡∏ô
‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢: XX.XX

[‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏Ñ‡∏ô]
```

````python
# ‡∏û‡∏¥‡∏°‡∏û‡πå‡πÇ‡∏Ñ‡πâ‡∏î‡∏Ñ‡∏≥‡∏ï‡∏≠‡∏ö‡∏ó‡∏µ‡πà‡∏ô‡∏µ‡πà

students = {
    "6601001": {"name": "‡∏™‡∏°‡∏ä‡∏≤‡∏¢", "score": 78},
    "6601002": {"name": "‡∏™‡∏°‡∏´‡∏ç‡∏¥‡∏á", "score": 85},
    "6601003": {"name": "‡∏™‡∏°‡∏´‡∏ß‡∏±‡∏á", "score": 92}
}
students.update({"6601004": {"name": "‡∏™‡∏°‡∏à‡∏¥‡∏ï", "score": 89}})
students["6601001"]["score"] = 80

print("--- ‡∏£‡∏≤‡∏¢‡∏ä‡∏∑‡πà‡∏≠‡∏ô‡∏±‡∏Å‡∏®‡∏∂‡∏Å‡∏©‡∏≤‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î ---")
for student_id, details in students.items():
    name = details["name"]
    score = details["score"]
    print(f"‡∏£‡∏´‡∏±‡∏™: {student_id}, ‡∏ä‡∏∑‡πà‡∏≠: {name}, ‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô: {score}")
print("----------------------------")

total_score = 0
student_count = 0

for student_id, details in students.items():
    total_score += details["score"]
    student_count += 1

average_score = total_score / student_count
print(f"‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢: {average_score}")
````

### ‡∏Ç‡πâ‡∏≠ 4: ‡∏ô‡∏¥‡∏û‡∏à‡∏ô‡πå‡∏ó‡∏≤‡∏á‡∏Ñ‡∏ì‡∏¥‡∏ï‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå (Mathematical Expressions)

‡πÉ‡∏´‡πâ‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô‡πÇ‡∏õ‡∏£‡πÅ‡∏Å‡∏£‡∏°‡πÅ‡∏õ‡∏•‡∏á‡∏ô‡∏¥‡∏û‡∏à‡∏ô‡πå‡∏ó‡∏≤‡∏á‡∏Ñ‡∏ì‡∏¥‡∏ï‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå‡∏ï‡πà‡∏≠‡πÑ‡∏õ‡∏ô‡∏µ‡πâ‡πÄ‡∏õ‡πá‡∏ô‡πÇ‡∏Ñ‡πâ‡∏î Python ‡πÅ‡∏•‡∏∞‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå:

1. **‡∏û‡∏∑‡πâ‡∏ô‡∏ó‡∏µ‡πà‡∏ß‡∏á‡∏Å‡∏•‡∏°:** A = œÄr¬≤ ‡πÇ‡∏î‡∏¢‡πÉ‡∏´‡πâ r = 5
2. **‡∏ô‡∏¥‡∏û‡∏à‡∏ô‡πå‡πÄ‡∏®‡∏©‡∏™‡πà‡∏ß‡∏ô:** (15 + 25) √ó 3 √∑ 2‚Å¥
3. **‡∏£‡∏≤‡∏Å‡∏ó‡∏µ‡πà 3:** ‚àõ64 (cube root of 64)
4. **‡∏£‡∏∞‡∏¢‡∏∞‡∏ó‡∏≤‡∏á:** d = ‚àö((x‚ÇÇ-x‚ÇÅ)¬≤ + (y‚ÇÇ-y‚ÇÅ)¬≤) ‡πÇ‡∏î‡∏¢‡πÉ‡∏´‡πâ (x‚ÇÅ,y‚ÇÅ) = (3,4) ‡πÅ‡∏•‡∏∞ (x‚ÇÇ,y‚ÇÇ) = (7,1)

**‡∏´‡∏°‡∏≤‡∏¢‡πÄ‡∏´‡∏ï‡∏∏:** ‡πÉ‡∏ä‡πâ `math` module ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö pi ‡πÅ‡∏•‡∏∞ sqrt

**Expected Output:**
```
1. ‡∏û‡∏∑‡πâ‡∏ô‡∏ó‡∏µ‡πà‡∏ß‡∏á‡∏Å‡∏•‡∏° = 78.50
2. ‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå‡∏ô‡∏¥‡∏û‡∏à‡∏ô‡πå = 7.50
3. ‡∏£‡∏≤‡∏Å‡∏ó‡∏µ‡πà 3 ‡∏Ç‡∏≠‡∏á 64 = 4.00
4. ‡∏£‡∏∞‡∏¢‡∏∞‡∏ó‡∏≤‡∏á = 5.00
```

````python
import math


r = 5
area = math.pi * r**2
formatted_area = f"{area:.2f}"

result_expr = (15 + 25) * 3 / 24
formatted_expr = f"{result_expr:.2f}"

value = 64
cube_root = value**(1/3)
formatted_root = f"{cube_root:.2f}"

x1, y1 = 3, 4
x2, y2 = 7, 1
distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
formatted_distance = f"{distance:.2f}"

print(f"1. ‡∏û‡∏∑‡πâ‡∏ô‡∏ó‡∏µ‡πà‡∏ß‡∏á‡∏Å‡∏•‡∏° = {formatted_area}")
print(f"2. ‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå‡∏ô‡∏¥‡∏û‡∏à‡∏ô‡πå = {formatted_expr}")
print(f"3. ‡∏£‡∏≤‡∏Å‡∏ó‡∏µ‡πà 3 ‡∏Ç‡∏≠‡∏á 64 = {formatted_root}")
print(f"4. ‡∏£‡∏∞‡∏¢‡∏∞‡∏ó‡∏≤‡∏á = {formatted_distance}")

# ‡∏û‡∏¥‡∏°‡∏û‡πå‡πÇ‡∏Ñ‡πâ‡∏î‡∏Ñ‡∏≥‡∏ï‡∏≠‡∏ö‡∏ó‡∏µ‡πà‡∏ô‡∏µ‡πà
````

‡πÉ‡∏´‡πâ‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô‡πÇ‡∏õ‡∏£‡πÅ‡∏Å‡∏£‡∏°‡πÄ‡∏û‡∏∑‡πà‡∏≠:
1. ‡∏£‡∏±‡∏ö‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô‡∏™‡∏≠‡∏ö‡∏Ç‡∏≠‡∏á‡∏ô‡∏±‡∏Å‡∏®‡∏∂‡∏Å‡∏©‡∏≤ 3 ‡∏ß‡∏¥‡∏ä‡∏≤

4. ‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡πÄ‡∏Å‡∏£‡∏î ‡πÇ‡∏î‡∏¢‡πÉ‡∏ä‡πâ‡πÄ‡∏á‡∏∑‡πà‡∏≠‡∏ô‡πÑ‡∏Ç‡∏ï‡πà‡∏≠‡πÑ‡∏õ‡∏ô‡∏µ‡πâ:
   - A: ‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô >= 80
   - B: ‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô >= 70

```python
# ‡πÉ‡∏´‡πâ‡∏£‡∏±‡∏ö‡∏Ñ‡πà‡∏≤‡∏à‡∏≤‡∏Å‡∏ú‡∏π‡πâ‡πÉ‡∏ä‡πâ‡∏´‡∏£‡∏∑‡∏≠‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡∏Ñ‡πà‡∏≤‡πÄ‡∏≠‡∏á
```

**Expected Output (‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏î‡πâ‡∏≤‡∏ô‡∏ö‡∏ô):**

```
‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô‡∏ß‡∏¥‡∏ä‡∏≤‡∏ó‡∏µ‡πà 3: 80
```

### ‡∏Ç‡πâ‡∏≠ 8: ‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏™‡∏ñ‡∏¥‡∏ï‡∏¥ (Function for Statistics)

‡πÉ‡∏´‡πâ‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô `calculate_stats(numbers)` ‡∏ó‡∏µ‡πà:
1. ‡∏£‡∏±‡∏ö list ‡∏Ç‡∏≠‡∏á‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç‡πÄ‡∏õ‡πá‡∏ô parameter
2. ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡πÅ‡∏•‡∏∞ return dictionary ‡∏ó‡∏µ‡πà‡∏°‡∏µ keys ‡∏ï‡πà‡∏≠‡πÑ‡∏õ‡∏ô‡∏µ‡πâ:
   - `min`: ‡∏Ñ‡πà‡∏≤‡∏ó‡∏µ‡πà‡∏ô‡πâ‡∏≠‡∏¢‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î
   - `max`: ‡∏Ñ‡πà‡∏≤‡∏ó‡∏µ‡πà‡∏°‡∏≤‡∏Å‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î
   - `sum`: ‡∏ú‡∏•‡∏£‡∏ß‡∏°‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î
   - `count`: ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç
   - `average`: ‡∏Ñ‡πà‡∏≤‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢ (rounded to 2 decimal places)

**‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏Å‡∏≤‡∏£‡πÉ‡∏ä‡πâ:**
```python
scores = [78, 85, 92, 64, 55, 88, 71, 95, 42, 80]
result = calculate_stats(scores)
print(result)
```

**Expected Output:**
```
{'min': 42, 'max': 95, 'sum': 750, 'count': 10, 'average': 75.0}
```

````python
# ‡∏û‡∏¥‡∏°‡∏û‡πå‡πÇ‡∏Ñ‡πâ‡∏î‡∏Ñ‡∏≥‡∏ï‡∏≠‡∏ö‡∏ó‡∏µ‡πà‡∏ô‡∏µ‡πà
# ‡πÉ‡∏´‡πâ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô calculate_stats ‡πÅ‡∏•‡πâ‡∏ß‡πÄ‡∏£‡∏µ‡∏¢‡∏Å‡πÉ‡∏ä‡πâ

scores = [78, 85, 92, 64, 55, 88, 71, 95, 42, 80]
def calculate_stats(numbers):
    if not numbers:
        return None
    
    stats = {
        'min': min(numbers),
        'max': max(numbers),
        'sum': sum(numbers),
        'count': len(numbers),
        'average': round(sum(numbers) / len(numbers), 2)
    }
    return stats

# ‡∏ó‡∏î‡∏™‡∏≠‡∏ö‡∏Å‡∏≤‡∏£‡πÉ‡∏ä‡πâ‡∏á‡∏≤‡∏ô
scores = [78, 85, 92, 64, 55, 88, 71, 95, 42, 80]
result = calculate_stats(scores)
print(result)

# result = calculate_stats(scores)
# print(result)
````

### ‡∏Ç‡πâ‡∏≠ 9: ‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏£‡∏´‡∏±‡∏™‡∏ú‡πà‡∏≤‡∏ô (Password Validation Function)

‡πÉ‡∏´‡πâ‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô `check_password(password)` ‡∏ó‡∏µ‡πà‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏Ñ‡∏ß‡∏≤‡∏°‡∏õ‡∏•‡∏≠‡∏î‡∏†‡∏±‡∏¢‡∏Ç‡∏≠‡∏á‡∏£‡∏´‡∏±‡∏™‡∏ú‡πà‡∏≤‡∏ô:

**‡πÄ‡∏á‡∏∑‡πà‡∏≠‡∏ô‡πÑ‡∏Ç‡∏Å‡∏≤‡∏£‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö:**
1. ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏¢‡∏≤‡∏ß‡∏ï‡πâ‡∏≠‡∏á‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏ô‡πâ‡∏≠‡∏¢ 8 ‡∏ï‡∏±‡∏ß‡∏≠‡∏±‡∏Å‡∏©‡∏£
2. ‡∏ï‡πâ‡∏≠‡∏á‡∏°‡∏µ‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç (0-9) ‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏ô‡πâ‡∏≠‡∏¢ 1 ‡∏ï‡∏±‡∏ß
3. ‡∏ï‡πâ‡∏≠‡∏á‡∏°‡∏µ‡∏ï‡∏±‡∏ß‡∏≠‡∏±‡∏Å‡∏©‡∏£‡∏û‡∏¥‡∏°‡∏û‡πå‡πÉ‡∏´‡∏ç‡πà (A-Z) ‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏ô‡πâ‡∏≠‡∏¢ 1 ‡∏ï‡∏±‡∏ß

**‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡∏à‡∏∞ return:**
- `True` ‡∏ñ‡πâ‡∏≤ password ‡∏õ‡∏•‡∏≠‡∏î‡∏†‡∏±‡∏¢
- `False` ‡∏ñ‡πâ‡∏≤ password ‡πÑ‡∏°‡πà‡∏õ‡∏•‡∏≠‡∏î‡∏†‡∏±‡∏¢ ‡∏û‡∏£‡πâ‡∏≠‡∏°‡∏Å‡∏±‡∏ö‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ö‡∏≠‡∏Å‡πÄ‡∏´‡∏ï‡∏∏‡∏ú‡∏•

**‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏Å‡∏≤‡∏£‡πÉ‡∏ä‡πâ:**
```python

check_password("hello")         # False
check_password("Hello12")       # False

```

````python
# ‡∏û‡∏¥‡∏°‡∏û‡πå‡πÇ‡∏Ñ‡πâ‡∏î‡∏Ñ‡∏≥‡∏ï‡∏≠‡∏ö‡∏ó‡∏µ‡πà‡∏ô‡∏µ‡πà
# ‡πÉ‡∏´‡πâ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô check_password ‡πÅ‡∏•‡πâ‡∏ß‡∏ó‡∏î‡∏™‡∏≠‡∏ö‡∏î‡πâ‡∏ß‡∏¢‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏ï‡πà‡∏≤‡∏á‡πÜ
import re

def check_password(password):
    if len(password) < 8:
        return False, "‡∏Ñ‡∏ß‡∏≤‡∏°‡∏¢‡∏≤‡∏ß‡∏ï‡πâ‡∏≠‡∏á‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏ô‡πâ‡∏≠‡∏¢ 8 ‡∏ï‡∏±‡∏ß‡∏≠‡∏±‡∏Å‡∏©‡∏£"
    
    if not re.search(r"\d", password):
        return False, "‡∏ï‡πâ‡∏≠‡∏á‡∏°‡∏µ‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç (0-9) ‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏ô‡πâ‡∏≠‡∏¢ 1 ‡∏ï‡∏±‡∏ß"
    
    if not re.search(r"[A-Z]", password):
        return False, "‡∏ï‡πâ‡∏≠‡∏á‡∏°‡∏µ‡∏ï‡∏±‡∏ß‡∏≠‡∏±‡∏Å‡∏©‡∏£‡∏û‡∏¥‡∏°‡∏û‡πå‡πÉ‡∏´‡∏ç‡πà (A-Z) ‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏ô‡πâ‡∏≠‡∏¢ 1 ‡∏ï‡∏±‡∏ß"
    return True, "‡∏£‡∏´‡∏±‡∏™‡∏ú‡πà‡∏≤‡∏ô‡∏õ‡∏•‡∏≠‡∏î‡∏†‡∏±‡∏¢"

print(check_password("Hello123"))
print(check_password("hello"))
print(check_password("Hello12"))
print(check_password("HELLO123"))

# test_passwords = ["Hello123", "hello", "Hello12", "HELLO123"]
# for pwd in test_passwords:
#     check_password(pwd)
````

### ‡∏Ç‡πâ‡∏≠ 10: ‡πÇ‡∏õ‡∏£‡πÅ‡∏Å‡∏£‡∏°‡∏à‡∏±‡∏î‡∏Å‡∏≤‡∏£‡∏£‡πâ‡∏≤‡∏ô‡∏Ñ‡πâ‡∏≤ (Inventory Management System)

**‡πÇ‡∏Ñ‡∏£‡∏á‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•:**
- Dictionary ‡∏ä‡∏∑‡πà‡∏≠ `inventory` ‡∏°‡∏µ structure: `{"‡∏ä‡∏∑‡πà‡∏≠‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤": {"price": ‡∏£‡∏≤‡∏Ñ‡∏≤, "quantity": ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô}}`

**‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡∏ó‡∏µ‡πà‡∏ï‡πâ‡∏≠‡∏á‡∏™‡∏£‡πâ‡∏≤‡∏á:**
1. `add_product(name, price, quantity)` - ‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤‡πÉ‡∏´‡∏°‡πà‡πÑ‡∏õ‡πÉ‡∏ô‡∏Ñ‡∏•‡∏±‡∏á‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤

3. `show_inventory()` - ‡πÅ‡∏™‡∏î‡∏á‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î ‡∏û‡∏£‡πâ‡∏≠‡∏° ‡∏£‡∏≤‡∏Ñ‡∏≤ ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô ‡∏°‡∏π‡∏•‡∏Ñ‡πà‡∏≤
4. `total_value()` - ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏°‡∏π‡∏•‡∏Ñ‡πà‡∏≤‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î (‡∏£‡∏≤‡∏Ñ‡∏≤ √ó ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô)

**‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏Å‡∏≤‡∏£‡πÉ‡∏ä‡πâ:**

```python
inventory = {}

print(f"‡∏°‡∏π‡∏•‡∏Ñ‡πà‡∏≤‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î: {total_value()} ‡∏ö‡∏≤‡∏ó")
```

```
   ‡∏£‡∏≤‡∏Ñ‡∏≤: 25 ‡∏ö‡∏≤‡∏ó/‡∏ä‡∏¥‡πâ‡∏ô

   ‡∏Ñ‡∏á‡πÄ‡∏´‡∏•‡∏∑‡∏≠: 20 ‡∏ä‡∏¥‡πâ‡∏ô
```

````python
# ‡∏õ‡∏£‡∏∞‡∏Å‡∏≤‡∏®‡∏ï‡∏±‡∏ß‡πÅ‡∏õ‡∏£ Global
inventory = {}

def add_product(name, price, quantity):
    if name in inventory:
        inventory[name]["quantity"] += quantity
    else:
        inventory[name] = {"price": price, "quantity": quantity}

def sell_product(name, quantity):
    # ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏ß‡πà‡∏≤‡∏°‡∏µ‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤‡πÅ‡∏•‡∏∞‡∏°‡∏µ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏û‡∏≠‡∏Ç‡∏≤‡∏¢‡∏´‡∏£‡∏∑‡∏≠‡πÑ‡∏°‡πà
    if name in inventory and inventory[name]["quantity"] >= quantity:
        inventory[name]["quantity"] -= quantity
    elif name not in inventory:
        print(f"--- ‡πÑ‡∏°‡πà‡∏û‡∏ö‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤: {name} ---")
    else:
        print(f"--- ‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤ {name} ‡πÑ‡∏°‡πà‡∏û‡∏≠‡∏Ç‡∏≤‡∏¢ ---")

def show_inventory():
    print("===== ‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤ =====")
    for i, (name, info) in enumerate(inventory.items(), 1):
        price = info["price"]
        qty = info["quantity"]
        value = price * qty
        print(f"{i}. {name}")
        print(f"   ‡∏£‡∏≤‡∏Ñ‡∏≤: {price} ‡∏ö‡∏≤‡∏ó/‡∏ä‡∏¥‡πâ‡∏ô")
        print(f"   ‡∏Ñ‡∏á‡πÄ‡∏´‡∏•‡∏∑‡∏≠: {qty} ‡∏ä‡∏¥‡πâ‡∏ô")
        print(f"   ‡∏°‡∏π‡∏•‡∏Ñ‡πà‡∏≤: {value} ‡∏ö‡∏≤‡∏ó")
        # ‡πÄ‡∏ß‡πâ‡∏ô‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î‡πÄ‡∏•‡πá‡∏Å‡∏ô‡πâ‡∏≠‡∏¢‡∏£‡∏∞‡∏´‡∏ß‡πà‡∏≤‡∏á‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£ ‡∏¢‡∏Å‡πÄ‡∏ß‡πâ‡∏ô‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£‡∏™‡∏∏‡∏î‡∏ó‡πâ‡∏≤‡∏¢
        if i < len(inventory):
            print("")
    print("\n========================")

def total_value():
    return sum(info["price"] * info["quantity"] for info in inventory.values())

# --- ‡∏ó‡∏î‡∏•‡∏≠‡∏á‡∏£‡∏±‡∏ô‡∏ï‡∏≤‡∏°‡πÇ‡∏à‡∏ó‡∏¢‡πå ---
inventory = {}

add_product("‡∏ô‡∏°", 25, 10)
add_product("‡∏Ç‡∏ô‡∏°‡∏õ‡∏±‡∏á", 15, 20)
add_product("‡πÑ‡∏Ç‡πà", 30, 5)

sell_product("‡∏ô‡∏°", 3)

show_inventory()
print(f"‡∏°‡∏π‡∏•‡∏Ñ‡πà‡∏≤‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î: {total_value()} ‡∏ö‡∏≤‡∏ó")
````
